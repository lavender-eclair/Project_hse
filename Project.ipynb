```python
import math
import numpy as np
import scipy.linalg as sla
from scipy import linalg



def what_curve(data):
    det_1 = data[0] + data[2]
    A = np.array([[data[0], data[1] / 2],
                  [data[1] / 2, data[2]]])
    det_2 = sla.det(A)
    B = np.array([[data[0], data[1] / 2, data[3] / 2],
                  [data[1] / 2, data[2], data[4] / 2],
                  [data[3] / 2, data[4] / 2, data[5]]])
    det_3 = sla.det(B)
    
    if (data[0] == 0 and data[1] == 0 and data[2] == 0):
        return "Уравнение не задает кривую второго порядка"
    
    else:
        if(det_2 > 0):
            if (det_3 == 0):
                return "Пара мнимых пересекающихся прямых"
            elif (det_1 * det_3 < 0):
                return "Эллипс"
            elif (det_1 * det_3 > 0):
                return "Мнимый эллипс"
        
        elif (det_2 < 0):
            if (det_3 != 0):
                return "Гипербола"
            elif (det_3 == 0):
                return "Пара пересекающихся прямых"
        elif (det_2 == 0):
            if (det_3 != 0):
                return "Парабола"
            elif (det_3 == 0):
                if (data[5] < 0):
                    return "Пара параллельных прямых"
                elif (data[5] > 0):
                    return "Пара мнимых параллельных прямых"
                elif (data[5] == 0):
                    return "Пара совпадающих прямых"
                
    return 0

def roots_of_characteristic_polynomial(data):
    A = np.array([[data[0], data[1] / 2],
                  [data[1] / 2, data[2]]])
    la, l = linalg.eig(A)
    l1, l2 = la
    return l1.real, l2.real

def canonical_form(v, data):
    if (data[1] == 0):
        phi = 0
    else:
        phi = (math.pi/2 - math.atan((data[0] - data[2]) / 2 )) / 2
    a = np.array([data[3] / 2, data[4] / 2])
    A = np.array([[math.cos(phi), -math.sin(phi)],
                 [math.sin(phi), math.cos(phi)]])
    b = a.dot(A)
    b_13 = b[0]
    b_23 = b[1]
    b_33 = data[5]
    if (v[0] != 0 and v[1] != 0):
        c = b_33 - b_13**2 / v[0]**2 - b_23**2/v[1]**2
        a_2 = c / v[0]
        b_2 = c / v[1]
        return (str(1 / a_2) + "*" + "x^2" + "+" + str(1 / b_2) + "*" + "y^2" + "+" + "1" + "=" + "0", (a_2, b_2))
    elif (v[0] == 0 and v[1] != 0):
        if (b_23 != 0):
            p = b_23 / v[1]
            c = b_33 - b_23**2 / v[1]**2
            return ("y^2" + "+" + str(2 * p) + "*" + "x" + "=" + "0", p)
        elif (b_23 == 0):
            c = b_33 - b_23**2 / v[1]**2
            return "y^2" + "+" + str(c / v[1]**2 ) + "=" + "0", 0
    return 0
        
def asymptotes_hyperbola(coef):
    y_1 = str(math.sqrt(coef[1] / coef[0])) + "*" + "x"
    y_2 = str(-(math.sqrt(coef[1] / coef[0]))) + "*" + "x"
    return y_1, y_2
    

def eccentricity_hyperbola(coef):
    e = math.sqrt(1 + coef[1] / coef[0])
    return e
    
def directris_hyperbola(coef,  e):
    x_1 = math.sqrt(coef[0]) / e
    x_2 = -math.sqrt(coef[0]) / e
    return x_1, x_2

def tangent_hyperbola(coef, x_0, y_0):
    if (((x_0)**2 / coef[0] - (y_0)**2 / coef[1]) != 1):
        return "Точка (x_0, y_0) не принадлежит гиперболе"
    else:
        tangent = str(x_0 / coef[0]) +  "*" + "x" + srt(-y_0 / coef[1]) + "*" + "y" + "=" + "1"
        return tangent
    return 0

def eccentricity_ellipse(coef):
    e = math.sqrt(1 - coef[1] / coef[0])
    return e

def dirictris_ellipse(coef, e):
    x_1 = math.sqrt(coef[0]) / e
    x_2 = -math.sqrt(coef[0]) / e
    return x_1, x_2

def square_ellipse(coef):
    s = math.pi * math.sqrt(coef[0]) * math.sqrt(coef[1])
    return s

def tangent_ellipse(coef, x_0, y_0):
    if (((x_0)**2 / coef[0] + (y_0)**2 / coef[1]) != 1):
        return "Точка (x_0, y_0) не принадлежит эллипсу"
    else:
        tangent = str(x_0 / coef[0]) +  "*" + "x" + srt(y_0 / coef[1]) + "*" + "y" + "=" + "1"
        return tangent
    return 0
    
def focus_parabola(coef):
    return "F" + "(" + str(coef / 2) + "," + "0" + ")"

def directris_parabola(coef):
    return "x" + "=" + str(coef / 2)

def tangent_parabola(coef, x_0, y_0):
    if ((y_0**2 - 2 * coef * x_0) != 0):
         return "Точка (x_0, y_0) не принадлежит параболе"
    else:
        tangent = str(y_0) + "*" + "y" + "=" + str(coef) + "*" + "(" + "x" +  "+" + str(x_0) + ")"
        return tangent
    
data = list(map(int, input().split()))
#ввод коэффициентов кривой в общем виде: a_11*x^2 + 2*a_12*xy + a_22*y^2 + 2*a_13*x + 2*a_23*y + a_33 = 0

if len(data) != 6:
    print("Некорректно введенные данные")
    exit(0)
else:
    v = roots_of_characteristic_polynomial(data)
    type = what_curve(data)
    if (type == "Уравнение не задает кривую второго порядка"):
        print(type)
        exit(0)
    else:
        print(type)
        answer = canonical_form(v, data)
        print(answer[0])
        
        if (type == "Гипербола"):
            coef = answer[1]
            print(asymptotes_hyperbola(coef))
            e = eccentricity_hyperbola(coef)
            print(e)
            print(directris_hyperbola(coef,  e))
            x_0, y_0 = map(int, input().split())
            print(tangent_hyperbola(coef, x_0, y_0))
            
        elif (type == "Эллипс"):
            coef = answer[1]
            e = eccentricity_ellipse(coef)
            print(e)
            print(dirictris_ellipse(coef,  e))
            print(square_ellipse(coef))
            x_0, y_0 = map(int, input().split())
            print(tangent_ellipse(coef, x_0, y_0))
            
        elif (type == "Парабола"):
            coef = answer[1]
            print(focus_parabola(coef))
            print(directris_parabola(coef))
            x_0, y_0 = map(int, input().split())
            print(tangent_parabola(coef, x_0, y_0))
```
